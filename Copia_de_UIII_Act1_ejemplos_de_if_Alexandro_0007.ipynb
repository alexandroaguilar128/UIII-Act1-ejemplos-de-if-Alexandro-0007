#### Ejemplo de if



```python
edad = 20

if edad >= 18:
    print("Eres mayor de edad.")
### Descripción de las líneas de código:
### edad = 20: Se define una variable llamada `edad` y se le asigna el valor numérico 20 [3].
### if edad >= 18: Es la sentencia condicional. Evalúa si el valor de la variable `edad` es mayor o igual a 18 utilizando un operador relacional [1, 2]. La línea termina obligatoriamente con dos puntos (`:`).
### print("Eres mayor de edad."): Este bloque de código está **sangrado** (identado), lo que indica que solo se ejecutará si la condición anterior resulta ser verdadera [1].
```

### Ejemplo if else

```python
# Ejemplo 2: Año bisiesto (Aritméticos, Lógicos y if/else)
# Este ejemplo utiliza el operador de módulo (%) y lógica compleja con and y or para determinar si un año es bisiesto
año = 2024                                 # Define el año a evaluar en la variable 'año' [18].

# Un año es bisiesto si es divisible por 4 y no por 100, O si es divisible por 400 [18].
if (año % 4 == 0 and año % 100 != 0) or (año % 400 == 0):
    # Línea anterior: '%' calcula el resto; '==' compara con 0; 'and'/'or' unen condiciones [18].
    print(f"{año} es bisiesto.")           # Se imprime si la lógica compleja resulta verdadera [18].
else:                                      # Se activa si la condición del 'if' es falsa [18].
    print(f"{año} no es bisiesto.")        # Mensaje para años no bisiestos [18].
```

### Ejemplo if elif

```python
# Ejemplo 3: Cálculo de descuentos (Aritméticos, Relacionales y elif)
# Calcula un descuento basado en la cantidad comprada, aplicando operaciones matemáticas al resultado final
cantidad = 12                              # Define la cantidad de productos comprados [17].

if cantidad >= 10:                         # Verifica si se compraron 10 o más artículos [17].
    descuento = 0.2                        # Asigna un 20% de descuento si la condición se cumple [17].
elif cantidad >= 5:                        # Si compró menos de 10 pero 5 o más [17].
    descuento = 0.1                        # Asigna un 10% de descuento [17].
else:                                      # Si compró menos de 5 artículos [17].
    descuento = 0                          # No se aplica ningún descuento [17].

total = 100 * (1 - descuento)              # Usa operadores aritméticos (*, -) para calcular el precio final [8].
print(f"Total a pagar: {total}")           # Muestra el monto final calculado tras las condiciones.
```

```python
# Ejemplo 4: Validación de rangos (Lógicos and y Relacionales)
# Verifica si un número se encuentra dentro de un rango específico
nota = 15                                  # Variable que almacena la calificación.

if nota >= 10 and nota <= 20:              # Usa 'and' para asegurar que se cumplan ambos límites [19].
                                           # También se puede escribir como: 10 <= nota <= 20 [19].
    print("La nota está en el rango permitido.") # Mensaje si el número está entre 10 y 20 inclusive.
else:                                      # Se ejecuta si el número está fuera del rango [19].
    print("La nota es inválida.")          # Mensaje de error si la condición 'and' falla [19].
```
